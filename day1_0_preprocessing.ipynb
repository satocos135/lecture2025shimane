# データ前処理


## RMecabの基礎

+ MeCab: 形態素解析エンジン
+ RMeCab: MeCabをR上で使うためのパッケージ


```r
# ライブラリの読み込み
library('dplyr')
library('RMeCab')
```

```r
# Rのバージョン
R.Version()$version.string
```

```r
# リスト形式で結果が返される
RMeCabC('今日は本を読んだ')
```

```r
# unlist()を使うとベクトルに変換できる
RMeCabC('今日は本を読んだ') %>% unlist()
```

```r
test = RMeCabC('今日は本を読んだ') %>% unlist()
```

```r
names(test)
```

```r
# 結果から名詞と動詞を抜き出す
# names()を使うとラベルを抜き出せる

test[names(test) %in% c('名詞', '動詞')] 
```

```r
names(test) %in% c('名詞', '動詞') 
```

```r
# 第二引数に1を入力すると、動詞の原形になる

RMeCabC('今日は本を読んだ', 1) %>% unlist() 
```

## クリーニングの実際

例文：

「やまいり？やまはいって木を丸める？」

**「切ってきて、丸めて、家の屋根にしげておくん。それはてんかごめんけどどこのやまいってきってきてもよいということになっちょった。」**


```r
text = '切ってきて、丸めて、家の屋根にしげておくん。それはてんかごめんけどどこのやまいってきってきてもよいということになっちょった。'
```

```r
RMeCabC(text)
```

表現を変える

'切ってきて、丸めて、家の屋根にしげておくん。それはてんかごめんけどどこのやまいってきってきてもよいということになっちょった。'

'切ってきて、丸めて、家の屋根に す げておくん。それは 天下御免 けどどこの山 に行って切ってきてもよいということになっていた。'

```r
text = '切ってきて、丸めて、家の屋根にすげておくん。それは天下御免でどこの山に行って切ってきてもよいということになっていた。'
```

```r
RMeCabC(text)
```

```r
# 辞書登録前
text = 'まあそういうことで、余分な話はいいけん。'
```

```r
RMeCabC(text)
```

```r
dict = 'dict/example_utf8.dic' # UTF-8版の場合
# dict = 'dict/example_sjis.dic' # Shift-JIS版の場合
```

```r
# 辞書登録後
RMeCabC(text, dic=dict) # utf-8版の場合
```

```r
# 辞書登録前
text = 'そこに来ちょる。'
RMeCabC(text)
```

```r
# 辞書登録後
RMeCabC(text, dic=dict)
```

```r
# 活用語は活用ごとに登録しないといけない
RMeCabC('そこに来ちょった。', dic=dict)
```

```r
text = 'しまあたのぉ、ここへ今日イノシシの肉が家にあったけん持ってくーだった'
```

```r
RMeCabC(text)
```

```r
text = 'しまったなぁ、ここへ今日イノシシの肉が家にあったから持ってくるんだった'
```

```r
RMeCabC(text)
```
